```python
import pandas as pd
import numpy as np
```

```python
json_file = "./arxiv-metadata-oai-snapshot.json"
data = pd.read_json(json_file, lines=True, encoding ='utf8')
data.head()
```

```python
data2_df = data[['authors','update_date','title','categories']]
data2_df.head()
```

```python
authors=[]
update_date=[]
categories=[]
title=[]
for i in range(1761194):
    authors.append(data2_df["authors"][i])
    update_date.append(data2_df["update_date"][i])
    categories.append(data2_df["categories"][i].split())
    title.append(data2_df["title"][i])
    if (i % 100000) ==0:
        print(str(i)+"mark reached")
condensed_arxiv = pd.DataFrame({"authors":authors,"update_date":update_date,"title":title,"categories":categories})
# condensed_arxiv.to_csv("arxiv_condensed_data.csv", index=True, header=True)
```
